4. Carga de una imagen y transformaciones

```python
import matplotlib.pyplot as plt
from skimage.transform import resize

# Ruta de la imagen
imagen_path = 'gato.PNG'

# Cargar la imagen
image = skimage.io.imread(imagen_path)

# Rotar la imagen 90 grados en sentido antihorario
rotated_image = skimage.transform.rotate(image, angle=90)

# Aplicar un filtro Gaussiano para suavizar la imagen
smoothed_image = skimage.filters.gaussian(image, sigma = 2)

image = resize(image, (300, 300))
rotated_image = resize(rotated_image, (300, 300))
smoothed_image = resize(smoothed_image, (300, 300))

import numpy as np
matriz_imagenes = np.array([image, rotated_image, smoothed_image])

#Mostrando la matriz de im치genes (si deseas visualizarlas)

fig, axs = plt.subplots(nrows=1, ncols=3, figsize = (20,20))
for i in range(3):
    axs[i].imshow(matriz_imagenes[i])
    axs[i].axis('off')  # Desactiva los ejes
plt.show()
```
