# Neuronal Network for Fahrenheit Grades Prediction

**Regular Programming**

```python
def function(c):
  f = c * 1.8 + 32
  return f

function(100)
```

**Automatic Programming**

```python
# Automatic Programming

import tensorflow as tf
import numpy as np
```

```python
# Input and output data
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype = float)
```

```python
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
# units: especifíca el número de neuronas en la capa
modelo = tf.keras.Sequential([capa])
```

```python
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(0.1), loss='mean_squared_error'
)
```

```python
print("Comenzando el entrenamiento...............")
historial= modelo.fit(celsius, fahrenheit, epochs=1000, verbose=False)
print("Modelo Entrenado")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("#Cantidad de vueltas..")
plt.ylabel("Magnitud de pérdida..")
plt.plot(historial.history["loss"])
```

```python
print("Hagamos una predicción..")
resultado = modelo.predict([100.0])
print("El resultado es: " + str(resultado) + "fharenheit")
```
